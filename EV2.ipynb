```python
import datetime 
import csv 
import openpyxl
import sys

ejemplares={}

while True: 
  opcion=int(input("*MENÚ PRINCIPAL*\nIngrese un numero de opcion a ejecutar:\n=Opción 1: Registrar un nuevo ejemplar\n=Opción 2: Consultas y reportes de un ejemplar\n=Opción 3: Salir del Programa\n> "))
  if opcion == 1:
    titulo=input("Ingrese el titulo del nuevo ejemplar:\n> ").upper()

    autor=input(f"Ingrese el nombre del autor del ejemplar {titulo}:\n> ").upper()

    genero=input(f"Ingrese el tipo de género que tiene el ejemplar:\n> ").upper()

    año_publicacion=input(f"Ingrese el año de publicación del ejemplar:\n> ")
    año=año_publicacion

    isbn=input("Ingrese el ISBN del ejemplar:\n> ").upper()

    fecha_adquisicion=input("Ingrese la fecha de adquisición del ejemplar con el siguiente formato(dd/mm/aaaa):\n> ")
     
    fecha=datetime.datetime.strptime(fecha_adquisicion, '%d/%m/%Y').date()
    clave=max(ejemplares, default=0)+1
    ejemplares[clave]=[titulo, autor, genero, año,isbn, fecha]
    
    with open("todos_los_ejemplares.csv","w", newline="") as archivo_csv:
      grabador_csv = csv.writer(archivo_csv)
      grabador_csv.writerow(("clave", "titulo", "autor", "genero", "año", "isbn", "fecha"))
      for clave, datos in ejemplares.items():
        grabador_csv.writerows((clave, datos[0], datos[1], datos[2], datos[3], datos[4], datos[5]))
      archivo_csv.close()
  if opcion == 2:
    while True:
      consultas_reportes=int(input("*MENÚ DE CONSULTAS Y REPORTES*\
      \nSeleccione una de las siguientes opciónes que desee ejecutar:\
      \n=Opción 1: Consultar titulo\n=Opción 2: Reporte\n=Opción 3: Volver al Menú Principal\n> "))
      if consultas_reportes == 1:
          consultas_titulo=int(input("Consultas de Titulo:\n-Opción 1: Consultar por medio de título:\
          \n-Opción 2: Volver al Menú de Consultas y Reportes\n> "))
          if consultas_titulo == 1:
            lista_titulo=[titulo for titulo, autor, genero, año, isbn, fecha in ejemplares.values()]
            print(f"Titulos de ejemplares existentes:")
            print(f"{lista_titulo}")
            consulta_reporte=input("Ingrese el titulo que desee consultar para mostrar sus datos:\n> ").upper()
            for titulo, autor, genero, año, isbn, fecha in ejemplares.values():
              if titulo == consulta_reporte:
                print(f"=Titulo: {titulo}\n=Autor: {autor}\n=Genero: {genero}\n=Año de Publicación:\
                 {año}\n=ISBN: {isbn}\n=Fecha: {fecha}")
          if consultas_titulo == 2:
            break

      if consultas_reportes == 2:
        while True:
          reportes=int(input("**MENÚ DE REPORTES**\nSeleccione el tipo de Reporte que desee:\
          \n=Opción 1: Catálogo completo\n=Opción 2: Reporte por Autor\
          \n=Opción 3: Reporte por Género\n=Opción 4: Reporte por Publicación de Año\
          \n=Opción 5: Volver al Menú de Consultas y Reportes\n>"))
          if reportes == 1:
            print(f"="* 30)
            print("**EJEMPLARES**")
            print(f"="* 30)
            for titulo, autor, genero, año, isbn, fecha in ejemplares.values():
              print(f"=Titulo: {titulo}\n=Autor: {autor}\n=Genero: {genero}\n=Año de Publicación: {año}\n=ISBN: {isbn}\n=Fecha: {fecha}")
              print(f"="* 30)
            while True:
              opcion_extraer=int(input("***MENU EXTRACCIONES***\
              \nSeleccione el tipo de exportación que desee: \
              \nOpción 1: Exportar en CSV\nOpción 2: Exportar en Excel\
              \nOpción 3: Volver\n> "))
              if opcion_extraer == 1: 
                with open("todos_los_ejemplares.csv","w", newline="") as archivo_csv:

                  grabador_csv = csv.writer(archivo_csv)

                  grabador_csv.writerow(("clave", "titulo", "autor", "genero", "año", "isbn", "fecha"))

                  for clave, datos in ejemplares.items():
                    grabador_csv.writerow((clave, datos[0], datos[1], datos[2], datos[3], datos[4], datos[5]))

                  archivo_csv.close()
              if opcion_extraer == 2:
                libro = openpyxl.Workbook()
                hoja = libro.active
                hoja.title = "Todos los ejemplares"

                hoja["A1"].value = "CLAVE"
                hoja["B1"].value = "TÍTULO"
                hoja["C1"].value = "AUTOR"
                hoja["D1"].value = "GÉNERO"
                hoja["E1"].value = "AÑO"
                hoja["F1"].value = "ISBN"
                hoja["G1"].value = "FECHA"

                from openpyxl.styles import Font
                celda_negritas = hoja["A1":"G1"]
                negritas = Font(bold = True)
                for fila in celda_negritas:
                  for celda in fila:
                    celda.font = negritas

                # Agregar los valores a las celdas
                filas = 2
                for clave, valores in ejemplares.items():
                    hoja.cell(row=filas, column=1).value = clave
                    hoja.cell(row=filas, column=2).value = valores[0]
                    hoja.cell(row=filas, column=3).value = valores[1]
                    hoja.cell(row=filas, column=4).value = valores[2]
                    hoja.cell(row=filas, column=5).value = valores[3]
                    hoja.cell(row=filas, column=6).value = valores[4]
                    hoja.cell(row=filas, column=7).value = valores[5]
                    filas += 1

                maxima_longitud = 0
                for valor in ejemplares.values():
                  for ejemplar in valor:
                    if len(ejemplar) > maxima_longitud:
                      maxima_longitud = len(ejemplar)

                hoja.column_dimensions["B"].width = maxima_longitud
                hoja.column_dimensions["G"].width = maxima_longitud

                # Guardar el archivo de Excel
                libro.save("todos_los_ejemplares.xlsx")
              if opcion_extraer == 3:
                break

          if reportes == 2:
            while True:
              opcion_extraer=int(input("***MENU EXTRACCIONES***\
              \nSeleccione el tipo de exportación que desee: \
              \nOpción 1: Exportar en CSV\nOpción 2: Exportar en Excel\
              \nOpción 3: Volver\n> "))
              if opcion_extraer == 1:
                datos_a_leer = dict() 
                try:
                  with open("todos_los_ejemplares.csv","r", newline="") as archivo:
                      lector = csv.reader(archivo)
                      next(lector)
                  
                      for clave, titulo, autor, genero, año, isbn, fecha in lector:
                        datos_a_leer[int(clave)] = (titulo, autor, genero, año, isbn, fecha)
                        
                except FileNotFoundError:
                  print("El archivo no se encontró, se procede a trabajar con un conjunto vacío")
                except Exception:
                  Excepcion = sys.exc_info()
                  print(f"Ocurrió un problema del tipo: {Excepcion[0]}")
                  print(f"Mensaje del error: {Excepcion[1]}")
                else:
                  lista_autores=[autor for titulo, autor, genero, año, isbn, fecha in datos_a_leer.values()]
                  autores_unicos=set(lista_autores)
                  print(f"Autores de ejemplares existentes:")
                  print(f"{autores_unicos}")
                  consulta_reporte=input("Ingrese el autor que desee consultar para mostrar sus datos:\n> ").upper()
                  print(f"="*30)
                  print(f"Ejemplares del autor {consulta_reporte}")
                  print(f"="*30)

                  ejemplares_por_autor=[]
                  for titulo, autor, genero, año, isbn, fecha in datos_a_leer.values():
                    if autor == consulta_reporte:
                        ejemplares_por_autor.append((titulo, autor, genero, año, isbn, fecha))
                        print(f"=Titulo: {titulo}\n=Autor: {autor}\n=Genero: {genero}\n=Año de Publicación: {año}\n=ISBN: {isbn}\n=Fecha: {fecha}")
                        print(f"-"*30)  
                
                  if not ejemplares_por_autor:
                    print(f"No se encontraron ejemplares del autor {consulta_reporte}")
                  else:
                      print("Su reporte a sido exportado correctamente")
                with open("autores_ejemplares.csv","w", newline="") as archivo_csv:

                  grabador_csv = csv.writer(archivo_csv)

                  grabador_csv.writerow(("titulo", "autor", "genero", "año", "isbn", "fecha"))

                  for fila in ejemplares_por_autor:
                    grabador_csv.writerow(fila)
                  archivo_csv.close()
                
              if opcion_extraer == 2:
                lista_autores=[autor for titulo, autor, genero, año, isbn, fecha in datos_a_leer.values()]
                autores_unicos=set(lista_autores)
                print(f"Autores de ejemplares existentes:")
                print(f"{autores_unicos}")
                consulta_reporte=input("Ingrese el autor que desee consultar para mostrar sus datos:\n> ").upper()
                print(f"="*30)
                print(f"Ejemplares del autor {consulta_reporte}")
                print(f"="*30)

                ejemplares_por_autor=[]
                for titulo, autor, genero, año, isbn, fecha in datos_a_leer.values():
                  if autor == consulta_reporte:
                    ejemplares_por_autor.append((titulo, autor, genero, año, isbn, fecha))
                    print(f"=Titulo: {titulo}\n=Autor: {autor}\n=Genero: {genero}\n=Año de Publicación: {año}\n=ISBN: {isbn}\n=Fecha: {fecha}")
                    print(f"-"*30)  
                    if not ejemplares_por_autor:
                      print(f"No se encontraron ejemplares del autor {consulta_reporte}")
                libro = openpyxl.Workbook()
                hoja = libro["Sheet"]
                hoja.title = f"Ejemplares de {consulta_reporte}"

                hoja["A1"].value = "TÍTULO"
                hoja["B1"].value = "AUTOR"
                hoja["C1"].value = "GÉNERO"
                hoja["D1"].value = "AÑO"
                hoja["E1"].value = "ISBN"
                hoja["F1"].value = "FECHA"

                for row_num, ejemplar in enumerate(ejemplares_por_autor, 2):
                  for col_num, valor in enumerate(ejemplar, 1):
                      hoja.cell(row=row_num, column=col_num, value=valor)

                nombre_archivo = f"ejemplares_por_autor_{consulta_reporte}.xlsx"
                libro.save(nombre_archivo)
                print(f"Archivo guardado como {nombre_archivo}")
              if opcion_extraer == 3:
                break
          if reportes == 3:
            while True:
              opcion_extraer = int(input("***MENU EXTRACCIONES***\
                        \nSeleccione el tipo de exportación que desee: \
                        \nOpción 1: Exportar en CSV\nOpción 2: Exportar en Excel\
                        \nOpción 3: Volver\n> "))
              
              if opcion_extraer == 1:
                  datos_a_leer = dict()
                  try:
                      with open("todos_los_ejemplares.csv", "r", newline="") as archivo:
                          lector = csv.reader(archivo)
                          next(lector)

                          for clave, titulo, autor, genero, año, isbn, fecha in lector:
                              datos_a_leer[int(clave)] = (titulo, autor, genero, año, isbn, fecha)

                  except FileNotFoundError:
                      print("El archivo no se encontró, se procede a trabajar con un conjunto vacío")
                  except Exception:
                      Excepcion = sys.exc_info()
                      print(f"Ocurrió un problema del tipo: {Excepcion[0]}")
                      print(f"Mensaje del error: {Excepcion[1]}")
                  else:
                      lista_generos = [genero for titulo, autor, genero, año, isbn, fecha in datos_a_leer.values()]
                      generos_unicos = set(lista_generos)
                      print(f"Géneros de ejemplares existentes:")
                      print(f"{generos_unicos}")
                      consulta_reporte = input("Ingrese el género que desee consultar para mostrar sus datos:\n> ").upper()
                      print(f"=" * 30)
                      print(f"Ejemplares del género {consulta_reporte}")
                      print(f"=" * 30)

                      ejemplares_por_genero = []
                      for titulo, autor, genero, año, isbn, fecha in datos_a_leer.values():
                          if genero == consulta_reporte:
                              ejemplares_por_genero.append((titulo, autor, genero, año, isbn, fecha))
                              print(f"=Titulo: {titulo}\n=Autor: {autor}\n=Género: {genero}\n=Año de Publicación: {año}\n=ISBN: {isbn}\n=Fecha: {fecha}")
                              print(f"-" * 30)

                      if not ejemplares_por_genero:
                          print(f"No se encontraron ejemplares del género {consulta_reporte}")
                      else:
                          print("Su reporte ha sido exportado correctamente")

                      with open("generos_ejemplares.csv", "w", newline="") as archivo_csv:
                          grabador_csv = csv.writer(archivo_csv)
                          grabador_csv.writerow(("titulo", "autor", "genero", "año", "isbn", "fecha"))

                          for fila in ejemplares_por_genero:
                              grabador_csv.writerow(fila)
                          archivo_csv.close()

              if opcion_extraer == 2:
                  lista_generos = [genero for titulo, autor, genero, año, isbn, fecha in datos_a_leer.values()]
                  generos_unicos = set(lista_generos)
                  print(f"Géneros de ejemplares existentes:")
                  print(f"{generos_unicos}")
                  consulta_reporte = input("Ingrese el género que desee consultar para mostrar sus datos:\n> ").upper()
                  print(f"=" * 30)
                  print(f"Ejemplares del género {consulta_reporte}")
                  print(f"=" * 30)
                  ejemplares_por_genero=[]
                  for titulo, autor, genero, año, isbn, fecha in datos_a_leer.values():
                    if genero == consulta_reporte:
                      ejemplares_por_genero.append((titulo, autor, genero, año, isbn, fecha))
                      print(f"=Titulo: {titulo}\n=Autor: {autor}\n=Genero: {genero}\n=Año de Publicación: {año}\n=ISBN: {isbn}\n=Fecha: {fecha}")
                      print(f"-"*30)  
                      if not ejemplares_por_genero:
                        print(f"No se encontraron ejemplares del autor {consulta_reporte}")
                  libro = openpyxl.Workbook()
                  hoja = libro["Sheet"]
                  hoja.title = f"Ejemplares de {consulta_reporte}"

                  hoja["A1"].value = "TÍTULO"
                  hoja["B1"].value = "AUTOR"
                  hoja["C1"].value = "GÉNERO"
                  hoja["D1"].value = "AÑO"
                  hoja["E1"].value = "ISBN"
                  hoja["F1"].value = "FECHA"

                  for row_num, ejemplar in enumerate(ejemplares_por_genero, 2):
                    for col_num, valor in enumerate(ejemplar, 1):
                        hoja.cell(row=row_num, column=col_num, value=valor)

                  nombre_archivo = f"ejemplares_por_autor_{consulta_reporte}.xlsx"
                  libro.save(nombre_archivo)
                  print(f"Archivo guardado como {nombre_archivo}")
              if opcion_extraer == 3:
                break
       
          if reportes == 4:
            while True:
              opcion_extraer=int(input("***MENU EXTRACCIONES***\
              \nSeleccione el tipo de exportación que desee: \
              \nOpción 1: Exportar en CSV\nOpción 2: Exportar en Excel\
              \nOpción 3: Volver\n> "))
              
              if opcion_extraer == 1:
                  datos_a_leer = dict() 
                  try:
                      with open("todos_los_ejemplares.csv","r", newline="") as archivo:
                          lector = csv.reader(archivo)
                          next(lector)

                          for clave, titulo, autor, genero, año, isbn, fecha in lector:
                              datos_a_leer[int(clave)] = (titulo, autor, genero, año, isbn, fecha)

                  except FileNotFoundError:
                      print("El archivo no se encontró, se procede a trabajar con un conjunto vacío")
                  except Exception:
                      Excepcion = sys.exc_info()
                      print(f"Ocurrió un problema del tipo: {Excepcion[0]}")
                      print(f"Mensaje del error: {Excepcion[1]}")
                  else:
                      lista_años = [año for titulo, autor, genero, año, isbn, fecha in datos_a_leer.values()]
                      años_unicos = set(lista_años)
                      print(f"Años de publicación de ejemplares existentes:")
                      print(f"{años_unicos}")
                      consulta_reporte = input("Ingrese el año que desee consultar para mostrar sus datos:\n> ")
                      print(f"="*30)
                      print(f"Ejemplares del año {consulta_reporte}")
                      print(f"="*30)

                      ejemplares_por_año = []
                      for titulo, autor, genero, año, isbn, fecha in datos_a_leer.values():
                          if año == consulta_reporte:
                              ejemplares_por_año.append((titulo, autor, genero, año, isbn, fecha))
                              print(f"=Titulo: {titulo}\n=Autor: {autor}\n=Género: {genero}\n=Año de Publicación: {año}\n=ISBN: {isbn}\n=Fecha: {fecha}")
                              print(f"-"*30)  

                      if not ejemplares_por_año:
                          print(f"No se encontraron ejemplares del año {consulta_reporte}")
                      else:
                          print("Su reporte ha sido exportado correctamente")

                      with open("años_ejemplares.csv","w", newline="") as archivo_csv:
                          grabador_csv = csv.writer(archivo_csv)
                          grabador_csv.writerow(("titulo", "autor", "genero", "año", "isbn", "fecha"))

                          for fila in ejemplares_por_año:
                              grabador_csv.writerow(fila)
                          archivo_csv.close()
              if opcion_extraer == 2:
                lista_anos = [ano for titulo, autor, genero, ano, isbn, fecha in datos_a_leer.values()]
                anos_unicos = set(lista_anos)
                print(f"Años de publicación existentes:")
                print(f"{anos_unicos}")
                consulta_reporte = input("Ingrese el año que desee consultar para mostrar sus datos:\n> ")
                print(f"=" * 30)
                print(f"Ejemplares publicados en el año {consulta_reporte}")
                print(f"=" * 30)
                ejemplares_por_ano=[]
                for titulo, autor, genero, ano, isbn, fecha in datos_a_leer.values():
                    if ano == consulta_reporte:
                        ejemplares_por_ano.append((titulo, autor, genero, ano, isbn, fecha))
                        print(f"=Titulo: {titulo}\n=Autor: {autor}\n=Género: {genero}\n=Año de Publicación: {ano}\n=ISBN: {isbn}\n=Fecha: {fecha}")
                        print(f"-"*30)  
                if not ejemplares_por_ano:
                    print(f"No se encontraron ejemplares publicados en el año {consulta_reporte}")

                libro = openpyxl.Workbook()
                hoja = libro["Sheet"]
                hoja.title = f"Ejemplares publicados en {consulta_reporte}"

                hoja["A1"].value = "TÍTULO"
                hoja["B1"].value = "AUTOR"
                hoja["C1"].value = "GÉNERO"
                hoja["D1"].value = "AÑO"
                hoja["E1"].value = "ISBN"
                hoja["F1"].value = "FECHA"

                for row_num, ejemplar in enumerate(ejemplares_por_ano, 2):
                    for col_num, valor in enumerate(ejemplar, 1):
                        hoja.cell(row=row_num, column=col_num, value=valor)

                nombre_archivo = f"ejemplares_por_ano_{consulta_reporte}.xlsx"
                libro.save(nombre_archivo)
                print(f"Archivo guardado como {nombre_archivo}")
          if reportes == 5:
            break
      
      if consultas_reportes == 3:
        break
  if opcion == 3:
    print("---Programa Finalizado, que tenga un lindo día---")
    break
```
